```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
ccf=pd.read_csv("creditcard.csv")
```

```python
ccf.info()
```

```python
ccf.head(15)
```

```python
ccf.describe()
```

```python
ccf.isnull().sum()
```

```python
ccf['Class'].value_counts()
```

```python
normal=ccf[ccf.Class==0]
```

```python
fraud=ccf[ccf.Class==1]
```

```python
print(normal.shape)
```

```python
print(fraud.shape)
```

```python
normal.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
ccf.groupby('Class').mean()
```

```python
normal_sample=normal.sample(n=492)
```

```python
new_ccf=pd.concat([normal_sample,fraud],axis=0)
```

```python
new_ccf.head(10)
```

```python
new_ccf['Class'].value_counts()
```

```python
new_ccf.groupby('Class').mean()
```

```python
X=new_ccf.drop(columns='Class',axis=1)
```

```python
Y=new_ccf['Class']
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
X_train_prediction=model.predict(X_train)
```

```python
training_data_acuracy=accuracy_score(X_train_prediction,Y_train)*100
```

```python
print(f"Training Data Accuracy: {training_data_acuracy}%")
```

```python
X_test_prediction=model.predict(X_test)
```

```python
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)*100
```

```python
print(f"Test Data Accuracy: {test_data_accuracy}%")
```
